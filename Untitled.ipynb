```python
import ast

def extract_names_from_column(df, column_name, top_n=None):
    def extract_names(cell):
        try:
            data = ast.literal_eval(cell)
            names = [item['name'] for item in data if 'name' in item]
            return names[:top_n] if top_n else names
        except:
            return []

    df[column_name] = df[column_name].apply(extract_names)
    return df
```

```python
movies.head()
```

```python
# For cast (top 3 only), overwrite the original column
movies = extract_names_from_column(movies, 'cast', top_n=3)

# For genres (all), overwrite
movies = extract_names_from_column(movies, 'genres')

# For keywords (all), overwrite
movies = extract_names_from_column(movies, 'keywords')
```

```python
import ast

def extract_director(crew_str):
    try:
        crew_list = ast.literal_eval(crew_str)
        for member in crew_list:
            if member.get('job') == 'Director':
                return member.get('name', '').replace(" ", "")
    except:
        return ''
    return ''
```

```python
movies['crew'] = movies['crew'].apply(extract_director)
```

```python
movies.head()
```

```python
 movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
 movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
 movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['genres'] + movies['crew'].apply(lambda x: [x])
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
